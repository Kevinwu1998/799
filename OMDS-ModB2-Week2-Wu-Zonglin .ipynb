```python
import matplotlib.pyplot as plt
print("Feature Histograms")
df.hist(figsize=(6, 6), bins=30)  # Adjust figure size and number of bins
plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()
```

```python
df['age'].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

We are working for a wealth management service that collects data on the age of its customers. The goal is to understand who our typical customers are in terms of their age and tailor our services accordingly.
Looking at the histogram, we might see a high peak around a certain age range 30-40 years old.This could suggest that our core customer base is composed mostly of middle-aged adults.

The histogram shows a tail that extends into the 50-60+ age group, we might notice that fewer older customers are present.Older customers could be less engaged in some types of services, possibly due to retirement, lifestyle changes, or a lack of interest in the current product offerings

```python
df['balance'].hist(bins=50)
plt.title("balance Distribution")
plt.xlabel("balance")
plt.ylabel("Frequency")
plt.show()
```

```python
df.boxplot(column=['balance'])
plt.show()
```

The tail with large balances represents a much smaller group, but they may have very different needs. You might want to offer premium products like investment portfolios, wealth management, or exclusive rewards for these customers. 

```python
plt.scatter(df['age'], df['balance'], alpha=0.5)
plt.title("Balance vs. Age")
plt.xlabel("Age")
plt.ylabel("Balance")
plt.show()
```

```python
import seaborn as sns
correlation_matrix = df[['age', 'balance', 'duration','campaign']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.show()
```

There is no correlation (or a very weak correlation) between age and balance, it suggests that age does not have a strong influence on how much money customers have in their accounts. Younger customers might have accumulated savings or balances through specific investment opportunities.This finding might also indicate that customers are not engaging with the bank's financial products

The Independence of Balance and Duration in Customer Behavior.This finding challenges the traditional assumption that customers who stay with a bank for a longer period will naturally accumulate more savings or assets. It suggests that loyalty or tenure with the bank isn't a strong driver of financial success.
